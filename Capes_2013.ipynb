## Descrição gráfica dos dados brutos

Mostra em amarelo os valores nulos

```
plt.figure(figsize=(16,8))
sns.heatmap(dados.isnull(), yticklabels = False, cmap = 'viridis', cbar= False)
```

tipo jurídico de instituição vs conceito capes

```
sns.countplot(x = 'CS_STATUS_JURIDICO', data = dados, hue = 'CD_CONCEITO_PROGRAMA', palette = 'RdBu_r')
plt.title('Tipo de instituição vs Conceito Capes')
plt.rcParams['figure.figsize'] = (15,7)
```

```
sns.countplot(x = 'CS_STATUS_JURIDICO', data = dados, hue = 'NM_REGIAO',palette = 'RdBu_r')
plt.title('CS_STATUS_JURIDICO vs NM_REGIAO')
plt.xlabel('CS_STATUS_JURIDICO')
plt.ylabel('Contagem')
plt.rcParams['figure.figsize'] = (7,7)
```

```
sns.countplot(x = 'NM_REGIAO', data = dados, hue = 'CD_CONCEITO_PROGRAMA', palette = 'RdBu_r')
plt.title('Região vs Conceito Capes')
plt.legend(loc='best')
```

```
sns.countplot(x = 'CS_STATUS_JURIDICO', data = dados, hue = 'NM_GRAU_PROGRAMA', palette = 'RdBu_r')
plt.title('CS_STATUS_JURIDICO vs NM_GRAU_PROGRAMA')
```

```
sns.countplot(x = 'CD_CONCEITO_PROGRAMA', data = dados, hue = 'NM_GRAU_PROGRAMA', palette = 'RdBu_r')
plt.title('GRAU PROGRAMA vs Conceito Capes')
```

```
fig, ax = plt.subplots()
ax.bar('CD_CONCEITO_PROGRAMA','NM_GRAU_PROGRAMA')
ax.annotate("Maior valor", 
            xy=(10, 12),
            xycoords='data',
            xytext=(11, 10),
            textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
plt.show()
```

tipos jurídico de instituição vs quantidade de instituições

```
dados['CS_STATUS_JURIDICO'].hist(bins = 7, color = 'darkred', alpha = 0.4)
plt.title('Tipo de instituição vs Quantidade de instituições')
```

```
plt.figure(figsize = (10, 5))
plt.title('CS_STATUS_JURIDICO X ANO_INICIO_PROGRAMA')
sns.boxplot(x = 'CS_STATUS_JURIDICO', y = 'ANO_INICIO_PROGRAMA', data = dados)
```

```
plt.figure(figsize = (10, 5))
plt.title('CS_STATUS_JURIDICO X CD_CONCEITO_PROGRAMA')
sns.boxplot(x = 'CS_STATUS_JURIDICO', y = 'CD_CONCEITO_PROGRAMA', data = dados)
```

```
dados['ANO_INICIO_PROGRAMA'].hist(bins = 30, color = 'darkred', alpha = 0.4)
plt.title('ANO_INICIO_PROGRAMAS')
plt.xlabel('ANO_INICIO_PROGRAMAS')
plt.ylabel('Contagem')
```

```
sns.countplot(x = 'CD_CONCEITO_PROGRAMA', data = dados, hue = 'NM_GRANDE_AREA_CONHECIMENTO', palette = 'RdBu_r')
plt.title('NM_GRANDE_AREA_CONHECIMENTO vs Conceito Capes')
```

```
sns.countplot(x = 'NM_REGIAO', data = dados, hue = 'NM_GRANDE_AREA_CONHECIMENTO', palette = 'RdBu_r')
plt.title('NM_GRANDE_AREA_CONHECIMENTO vs Conceito Capes')
```
